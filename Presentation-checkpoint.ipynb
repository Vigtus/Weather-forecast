## Import danych z posortowanego pliku oraz wyswietlenie 5 pierwszych wierszy

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import pandas as pd
data = pd.read_csv("DataSets/seattle-weather.csv")
data['date'] = pd.to_datetime(data['date'])

print(data.head())
```

# Przystosowanie danych do uczenia modelu

## Usunięcie danych które zbyt odstają od badanego zakresu

```python
# Usuwanie wartości odstających
def remove_outliers(column):
    q1 = column.quantile(0.2)
    q3 = column.quantile(0.8)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return column[(column >= lower_bound) & (column <= upper_bound)]

for column in data.select_dtypes(include=['number']).columns:
    data[column] = remove_outliers(data[column])
```

## Usunięcie wierszy posiadające brakujące wartości z df

```python
data = data.dropna()
```

## Test stacjonarności 

```python
# Sprawdzenie stacjonarności 
from statsmodels.tsa.stattools import adfuller
def stationarity_test(column):
    x = adfuller(column, autolag='AIC')
    p = x[1]
    print(p)
    if(p<0.05):
        print("Stationary")
    else:
        print('Non-stationary')


def weekly_data(data):
    # Ustawienie kolumny daty jako indeks
    data.set_index('date', inplace=True)
    
    # Grupowanie danych co tydzień i obliczanie średniej temperatury
    weekly_data = data.resample('W').mean()
    
    return weekly_data

weekly_df = weekly_data(data[['date', 'temp_min','temp_max']])
stationarity_test(weekly_df['temp_min'])
```

## Wykresy ACF
Jest kolejnym narzędziem diagnostycznym używanym w analizie szeregów czasowych do identyfikacji wzorców autokorelacji w danych.ACF mierzy korelację między obserwacją w danym punkcie czasowym, a obserwacjami w poprzednich punktach czasowych.

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
# Wykres ACF
max_lags = 10  # Przykładowa wartość, możesz dostosować ją według potrzeb
plt.figure(figsize=(10, 5))  # Opcjonalnie, ustaw rozmiar wykresu
plot_acf(weekly_df['temp_min'], lags=max_lags)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.show()
```

## Wykres PACF
To narzędzie diagnostyczne używane w analizie szeregów czasowych do identyfikacji sezonowych lub innych nieliniowych wzorców w danych. PACF mierzy korelację między dwoma punktami w szeregu czasowym po usunięciu wpływu wszystkich wcześniejszych punktów w szeregu czasowym.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(10, 5))  # Opcjonalnie, ustaw rozmiar wykresu
plot_pacf(weekly_df['temp_min'], lags=max_lags)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

# Diagnostyka modelu Sarima

**Standardized Residuals Plot**
Ten wykres pokazuje znormalizowane reszty (błędy modelu) w funkcji czasu. Idealnie, powinny one być losowo rozproszone wokół zera bez jakichkolwiek widocznych wzorców. Jeśli widzisz jakieś wzorce, może to oznaczać, że model nie uwzględnia pewnych charakterystyk danych.

**Histogram plus KDE**
Pokazuje histogram znormalizowanych reszt (błędów) wraz z oszacowanym gęstościowym estymatorem jądrowym (KDE). Idealnie, powinien przypominać rozkład normalny, co oznacza, że reszty są losowo rozłożone wokół zera.

**Normal Q-Q Plot**
Jest to wykres kwantylowy kwantyla normalnego dla znormalizowanych reszt. Jeśli punkty na tym wykresie leżą na linii prostej, oznacza to, że reszty mają rozkład normalny.

**Correlogram**
Ten wykres pokazuje autokorelację reszt, co pozwala na sprawdzenie, czy istnieją jakiekolwiek wzory w resztach modelu. Idealnie, nie powinno być istotnych autokorelacji.

```python
# Plot diagnostics
plt.figure(figsize=(10, 8))
sarima_fit_result_temp_min.plot_diagnostics()
plt.show()
```

# Sprawdzanie efektywnosći/jakości między modelami

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define SARIMA models with different parameters
sarima_model1 = SARIMAX(train_data['temp_min'], order=(2, 0, 1), seasonal_order=(1, 1, 1, 52))
sarima_model2 = SARIMAX(train_data['temp_min'], order=(3, 0, 2), seasonal_order=(1, 1, 1, 52))

# Fit the SARIMA models
sarima_fit_result1 = sarima_model1.fit()
sarima_fit_result2 = sarima_model2.fit()

# Forecasting using SARIMA models
forecast_temp_min1 = sarima_fit_result1.get_forecast(steps=len(test_data))
forecast_temp_min2 = sarima_fit_result2.get_forecast(steps=len(test_data))

# Plotting
plt.figure(figsize=(10, 5))

# Plot SARIMA Model 1
plt.plot(train_data.index, train_data['temp_min'], label='Training Data', color='blue')
plt.plot(test_data.index, test_data['temp_min'], label='Test Data', color='green')
plt.plot(test_data.index, forecast_temp_min1.predicted_mean, label='SARIMA Model 1 Forecast', color='red')

# Plot SARIMA Model 2
plt.plot(test_data.index, forecast_temp_min2.predicted_mean, label='SARIMA Model 2 Forecast', color='orange')


plt.title('Comparison of SARIMA Models for Temperature Min')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

## Obliczenie który z modeli jest lepszy 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate forecast errors for SARIMA Model 1
forecast_errors1 = test_data['temp_min'] - forecast_temp_min1.predicted_mean
mse1 = np.mean(forecast_errors1 ** 2)
mae1 = np.mean(np.abs(forecast_errors1))
rmse1 = np.sqrt(mse1)

# Calculate forecast errors for SARIMA Model 2
forecast_errors2 = test_data['temp_min'] - forecast_temp_min2.predicted_mean
mse2 = np.mean(forecast_errors2 ** 2)
mae2 = np.mean(np.abs(forecast_errors2))
rmse2 = np.sqrt(mse2)

print("SARIMA Model 1:")
print("  MSE:", mse1)
print("  MAE:", mae1)
print("  RMSE:", rmse1)

print("\nSARIMA Model 2:")
print("  MSE:", mse2)
print("  MAE:", mae2)
print("  RMSE:", rmse2)

# Calculate the percentage improvement of Model 1 over Model 2
percentage_improvement = ((mse2 - mse1) / mse2) * 100

if mse1 < mse2:
    print("SARIMA Model 1 is better than SARIMA Model 2.")
    print("The MSE of Model 1 is {:.2f}% lower than Model 2.".format(percentage_improvement))
elif mse2 < mse1:
    print("SARIMA Model 2 is better than SARIMA Model 1.")
    print("The MSE of Model 2 is {:.2f}% lower than Model 1.".format(-percentage_improvement))
else:
    print("Both SARIMA models perform equally.")
```
